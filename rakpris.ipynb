```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#  Data fetched from http://www.r√§kpris.se/data
df = pd.read_csv("rakpris.csv", encoding='utf-8', header=0)
pd.created_at = pd.to_datetime(df.date, 
                               format='%Y%m%d')
df.index = pd.created_at
df.head()
```

```python
df.index
```

```python
def price(df):
    fig = plt.figure(figsize=(15,7))
    ax = fig.add_subplot(111)
    plt.plot(df.customerprice)
    plt.plot(df.auctionprice)
    plt.legend()
    plt.show()
price(df)
```

```python
def boxes(df):
    fig = plt.figure(figsize=(15,7))
    ax = fig.add_subplot(111)
    plt.plot(df.boxes)
    plt.legend()
    plt.show()
boxes(df)
```

```python
def kilo(df):
    fig = plt.figure(figsize=(15,7))
    ax = fig.add_subplot(111)
    plt.plot(df.kilo)
    plt.legend()
    plt.show()
kilo(df)
```
